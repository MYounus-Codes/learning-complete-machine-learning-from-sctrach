```python
c = np.array([[1, 2, 3], [4, 5, 6]]) # 2D array with shape (2, 3)
```

```python
print(c.shape) # (2, 3) 2 = number of rows, 3 = number of columns
print(c.ndim) # 2 (number of dimensions)
print(c.size) # 6 (total number of elements)
print(c.dtype) # int64 (data type)
```

```python
c = c.repeat(10, axis=0) # repeat each row 10 times
print(c)
```

```python
d = np.zeros((3, 4)) # 2D array with shape (3, 4) filled with zeros
print(d)
```

```python
e = np.ones((3, 6)) # 2D array with shape (2, 3) filled with ones
print(e)
```

```python
f = np.full((3, 4), 7) # 2D array with shape (3, 4) filled with sevens
print(f)
```

```python
g = np.eye(3, 5) # 2D array with shape (3, 5) filled with identity matrix
print(g)
```

```python
h = np.random.rand(3, 4) # 2D array with shape (3, 4) filled with random numbers between 0 and 1
print(h)
```

```python
i = np.random.randint(0, 10, (3, 4)) # 2D array with shape (3, 4) filled with random integers between 0 and 9
print(i)
```

```python
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])

# Opreations

matrix_c = matrix_a + matrix_b # element-wise addition
print(matrix_c)
matrix_d = matrix_a - matrix_b # element-wise subtraction
print(matrix_d)
matrix_e = matrix_a * matrix_b # element-wise multiplication
print(matrix_e)
```

```python
# Dot product
matrix_f = np.dot(matrix_a, matrix_b) # dot product
print(matrix_f)
```
