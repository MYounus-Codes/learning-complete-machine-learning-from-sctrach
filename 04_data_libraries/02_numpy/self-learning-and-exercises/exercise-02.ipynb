```python
import numpy as np

# Column 0 → Sales
# Column 1 → Advertising Cost
# Column 2 → New Customers
# Column 3 → Profit

df = np.array([
    [12000, 2000, 150, 3200],
    [15000, 2500, 180, 4200],
    [17000, 3000, 210, 5100],
    [16000, 2800, 205, 4800],
    [19000, 3500, 240, 6200],
    [21000, 3800, 260, 7000],
    [23000, 4000, 290, 7800],
    [22000, 3900, 275, 7400],
    [24000, 4200, 300, 8200],
    [26000, 4500, 330, 9100],
    [28000, 4700, 350, 9800],
    [30000, 5000, 380, 11000]
])
```

```python
# Print the minimum profit value.

np.min(df[:, 3])
```

```python
# Print the average sales value.

np.average(df[:, 0])
```

```python
# Print the total profit value.

np.sum(df[:, -1])
```

```python
# Find the month with the highest sales.

highest_sales_month = np.argmax(df[:, 0])
highest_sales_month
```

```python
# Find the month with the lowest advertising cost.

np.argmin(df[:, 1])
```

```python
# Print the average number of new customers.

np.average(df[:, 2])
```

```python
# Find the sum of each column

print(np.sum(df, axis=0))

# Find the sum of each row


print(np.sum(df, axis=1))
```

```python
# Find the average of each column

print(np.average(df, axis=0))
```

```python
# Increase all sales by 10% 

df[:, 0] = df[:, 0] * 1.10
print(df[:, 0])
```

```python
# Increase all advertising costs by 5%

df[:, 1] = df[:, 1] * 1.05
print(df[:, 1])
```

```python
# Print only the months where Profit > 7000.

print(df[df[:, 3] > 7000])
```
