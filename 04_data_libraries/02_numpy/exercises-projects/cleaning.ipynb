```python
df["Gender"] = df["Gender"].map({"Male":0,"Female":1})
df["Discount_Used"] = df["Discount_Used"].map({"No":0,"Yes":1})
df["Subscription"] = df["Subscription"].map({"No":0,"Yes":1})

city_mapping = {
    city:i
    for i,city in enumerate(df["City"].unique())
}

df["City"] = df["City"].map(city_mapping)

df.head(5)
```

```python
# Exporting the cleaned data to a new CSV file
df.to_csv("cleaned_business_sales.csv", index=False)
```
