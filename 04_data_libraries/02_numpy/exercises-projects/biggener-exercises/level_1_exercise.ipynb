```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../cleaned_business_sales.csv")
df.head()
```

```python
## Converting the df in to the numpy array
data = df.to_numpy()
```

## **Each Code Cell Solved One Excercise One BY One**

```python
## Print the shape of the dataset using the numpy

data.shape
```

```python
# Print number of rows and columns in the dataset using the numpy 

print("Number of rows:", data.shape[0])
print("Number of columns:", data.shape[1])
```

```python
# Print the first and last row

print("First row:", data[0])
print("Last row:", data[-1])
```

```python
print("Monthly_Income:\n", data[:, 4])
```

```python
print("Orders:\n", data[:, 5])
```

```python
## Print First five rows of the dataset using the numpy
print("First five rows:\n", data[:5])
```

```python
## Print Last five rows of the dataset using the numpy
print("Last five rows:\n", data[-5:])
```

```python
## Print Age and Monthly_Income columns together

print("Age and Monthly_Income columns:\n", data[:, [1, 4]])
```

```python
## Print customers whose age is greater than 30 
print("Customers with age greater than 30:\n", data[data[:, 1] > 30])
```

```python
## Print customers from Karachi

print("Customers from Karachi:\n", data[data[:, 3] == 0])
```

```python
## Print subscribed customers

print("Subscribed customers:\n", data[data[:, 10] == 1])
```

```python
## Print customers with monthly income greater than 80, 000
print("Customers with monthly income greater than 80,000:\n", data[data[:, 4] > 80000])
```

```python
## Print customers have more than 20 orders

print("Customers with more than 20 orders:\n", data[data[:, 5] > 20])
```
