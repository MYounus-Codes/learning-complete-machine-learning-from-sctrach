##### **Array Creation Ways**

```python
## Creating Arrays
import numpy as np

np.array() # Create an array from a list or tuple

np.zeros() # Create an array of zeros

np.ones() # Create an array of ones

np.empty() # Create an empty array (uninitialized values)

np.full() # Create an array filled with a specified value

np.arange() # Create an array with a range of values

np.linspace() # Create an array with evenly spaced values over a specified interval

np.eye() # Create a 2-D array with ones on the diagonal and zeros elsewhere

np.identity() # Create a square identity matrix
```

##### **Array Attributes**

```python
## Array Attributes

np.ndim # Number of dimensions of the array

np.shape # Shape of the array (tuple of dimensions)

np.size # Total number of elements in the array

np.dtype # Data type of the array elements

np.itemsize # Size in bytes of each element in the array

np.nbytes # Total size in bytes of the array
```

#### **Array Practice**

```python
# 1D Array Creation and Manipulation
import numpy as np

# Create a 1D array
array_1d = np.array([1, 2, 3, 4, 5])
print("Original 1D Array:", array_1d)

# Accessing elements
print("First element:", array_1d[0])

# Modifying elements
array_1d[2] = 10
print("Modified 1D Array:", array_1d)

# Array Slicing
sliced_array = array_1d[1:4]
print("Sliced Array (elements 1 to 3):", sliced_array)
```
