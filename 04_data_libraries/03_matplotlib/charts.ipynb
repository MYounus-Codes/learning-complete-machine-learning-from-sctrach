#### **Bar Chart**

```python
import matplotlib.pyplot as plt
```

```python
product = ['A', 'B', 'C', 'D']
sales = [100, 200, 300, 400]

plt.bar(product, sales, color='orange', label='Sales Data')
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
product = ['A', 'B', 'C', 'D']
sales = [100, 200, 300, 400]

plt.bar(product, sales, color=['red', 'blue', 'green', 'orange'], edgecolor='black', linewidth=1.5, alpha=0.7, linestyle='--', hatch='/', width=0.5, label='Sales Data')
plt.title('Sales by Product', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Product', fontsize=14, fontweight='bold', color='darkgreen')
plt.ylabel('Sales', fontsize=14, fontweight='bold', color='darkgreen')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=12, fontweight='bold', color='darkred')
plt.yticks(fontsize=12, fontweight='bold', color='darkred')
# Sending Grid line behind the bars
plt.gca().set_axisbelow(True)    
plt.legend()
plt.show()
```

```python
# Plotting Horizontal Bar Chart
plt.barh(product, sales, color='orange', label='Sales Data')
plt.title('Sales by Product')
plt.xlabel('Sales')
plt.ylabel('Product')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Define the dataset
categories = ['Python', 'Java', 'JavaScript', 'C++', 'Go']
values = [85, 72, 78, 65, 50]
colors = ['#3776AB', '#007396', '#F7DF1E', '#00599C', '#00ADD8'] # Custom brand hex colors

# 2. Create the figure and axis object (Recommended object-oriented style)
fig, ax = plt.subplots(figsize=(8, 5))

# 3. Plot the bar chart
bars = ax.bar(categories, values, color=colors, edgecolor='black', width=0.6)

# 4. Add data value labels on top of each bar automatically
ax.bar_label(bars, padding=3, fontsize=10, weight='bold')

# 5. Customize titles and axis labels
ax.set_title('Programming Language Popularity Index', fontsize=14, weight='bold', pad=15)
ax.set_xlabel('Languages', fontsize=12)
ax.set_ylabel('Score / Popularity (%)', fontsize=12)

# 6. Enhance visual aesthetics
ax.set_ylim(0, 100)                     # Set limits for y-axis to provide breathing room
ax.grid(axis='y', linestyle='--', alpha=0.7) # Add a subtle horizontal grid background
ax.set_axisbelow(True)                  # Push grid lines behind the bars

# 7. Render and save the graph
plt.tight_layout() # Prevents clipping of text elements
plt.show()
```
