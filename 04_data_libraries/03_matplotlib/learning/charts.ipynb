#### **Bar Chart**

```python
import matplotlib.pyplot as plt
```

```python
product = ['A', 'B', 'C', 'D']
sales = [100, 200, 300, 400]

plt.bar(product, sales, color='orange', label='Sales Data')
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
product = ['A', 'B', 'C', 'D']
sales = [100, 200, 300, 400]

plt.bar(product, sales, color=['red', 'blue', 'green', 'orange'], edgecolor='black', linewidth=1.5, alpha=0.7, linestyle='--', hatch='/', width=0.5, label='Sales Data')
plt.title('Sales by Product', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Product', fontsize=14, fontweight='bold', color='darkgreen')
plt.ylabel('Sales', fontsize=14, fontweight='bold', color='darkgreen')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=12, fontweight='bold', color='darkred')
plt.yticks(fontsize=12, fontweight='bold', color='darkred')
# Sending Grid line behind the bars
plt.gca().set_axisbelow(True)    
plt.legend()
plt.show()
```

```python
# Plotting Horizontal Bar Chart
plt.barh(product, sales, color='orange', label='Sales Data')
plt.title('Sales by Product')
plt.xlabel('Sales')
plt.ylabel('Product')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Define the dataset
categories = ['Python', 'Java', 'JavaScript', 'C++', 'Go']
values = [85, 72, 78, 65, 50]
colors = ['#3776AB', '#007396', '#F7DF1E', '#00599C', '#00ADD8'] # Custom brand hex colors

# 2. Create the figure and axis object (Recommended object-oriented style)
fig, ax = plt.subplots(figsize=(8, 5))

# 3. Plot the bar chart
bars = ax.bar(categories, values, color=colors, edgecolor='black', width=0.6)

# 4. Add data value labels on top of each bar automatically
ax.bar_label(bars, padding=3, fontsize=10, weight='bold')

# 5. Customize titles and axis labels
ax.set_title('Programming Language Popularity Index', fontsize=14, weight='bold', pad=15)
ax.set_xlabel('Languages', fontsize=12)
ax.set_ylabel('Score / Popularity (%)', fontsize=12)

# 6. Enhance visual aesthetics
ax.set_ylim(0, 100)                     # Set limits for y-axis to provide breathing room
ax.grid(axis='y', linestyle='--', alpha=0.7) # Add a subtle horizontal grid background
ax.set_axisbelow(True)                  # Push grid lines behind the bars

# 7. Render and save the graph
plt.tight_layout() # Prevents clipping of text elements
plt.show()
```

#### **Pie Chart**

```python
import matplotlib.pyplot as plt

cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
temperatures = [30, 25, 20, 35, 40]

plt.pie(temperatures, labels=cities, autopct='%1.1f%%', startangle=140, colors=['#FF9999','#66B3FF','#99FF99','#FFCC99','#FFD700'])
plt.title('Average Temperatures in Major US Cities', fontsize=14, fontweight='bold', pad=25)
plt.axis('equal')
plt.show()
```

```python
## Customizing Pie Chart with Exploded Slices and Shadow
import matplotlib.pyplot as plt

cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
temperatures = [30, 25, 20, 35, 40]

explode = (0.1, 0, 0, 0, 0)  # Explode the first slice (New York)
plt.pie(temperatures, labels=cities, autopct='%1.1f%%', startangle=140, colors=['#FF9999','#66B3FF','#99FF99','#FFCC99','#FFD700'], explode=explode, shadow=True)
plt.title('Average Temperatures in Major US Cities', fontsize=14, fontweight='bold', pad=25)
plt.axis('equal')
plt.show()
```

```python
## Editing the Pie Chart to Include a Legend and Custom Font Sizes, Using the Different Data
import matplotlib.pyplot as plt

ft_players = ['C. Ronaldo', 'Neymar', 'Messi', 'Mbappe', 'Haaland']
goals = [974, 461, 919, 420, 361] ## ALL Time Goals

explode = (0.1, 0, 0, 0, 0)

wedges, labels, autopct = plt.pie(goals, labels=ft_players, autopct='%1.1f%%', startangle=140, colors=['#FF9999','#66B3FF','#99FF99','#FFCC99','#FFD700'], explode=explode, shadow=True)

plt.title('Goal Scoring by Top Football Players', fontsize=14, fontweight='bold', pad=25)
plt.axis('equal')

# Add a legend with custom font size
plt.legend(wedges, ft_players, title="FT Players", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt

ft_players = ['C. Ronaldo', 'Neymar', 'Messi', 'Mbappe', 'Haaland']
goals = [974, 461, 919, 420, 361] ## ALL Time Goals

explode = (0.1, 0, 0, 0, 0)
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#FFD700']

# Create figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Generate the pie chart
wedges, labels, autopct = ax.pie(
    goals, 
    labels=ft_players, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors, 
    explode=explode, 
    shadow=True,
    textprops={'fontsize': 11, 'color': 'black'} # Sets font size for labels and percentages
)

# Customize percentage font styles specifically (if you want them bolder or a different color)
for pct in autopct:
    pct.set_fontsize(10)
    pct.set_fontweight('bold')

ax.set_title('Goal Scoring by Top Football Players', fontsize=16, fontweight='bold', pad=25)
ax.axis('equal')

# Add a legend with custom font size
ax.legend(
    wedges, 
    ft_players, 
    title="FT Players", 
    loc="center left", 
    bbox_to_anchor=(1.1, 0.5), 
    fontsize=12,
    title_fontsize=13
)

# Adjust layout so the legend doesn't get cut off when exporting or displaying
plt.tight_layout()

plt.show()
```

#### **Histogram**

```python
import matplotlib.pyplot as plt

## Creadting a DataSet for the Histogram: Marks of students in a class, out of 100
marks = [20, 60, 11, 45, 90, 32, 88, 40, 70, 82, 80, 40, 60, 75, 30]
plt.hist(marks, bins=5, color='green', edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Distribution of Student Marks')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating a DataSet for the Histogram: Marks of students in a class, out of 100
marks = [20, 60, 11, 45, 90, 32, 88, 40, 70, 82, 80, 40, 60, 75, 30]

# 1. Enhance visual styling using a clean background style
plt.style.use('seaborn-v0_8-whitegrid')

# 2. Set an explicit figure size for better readability
plt.figure(figsize=(9, 5.5))

# 3. Use explicit bin edges (0 to 100 with steps of 20) instead of a raw integer
#    This forces bins to map perfectly to standard grade boundaries (e.g., 0-20, 20-40...)
bin_edges = np.arange(0, 101, 20)

# 4. Modernize colors (Steel blue fills with slightly lighter edges to prevent visual clutter)
#    Added rwidth to create elegant spacing between bars
counts, bins, patches = plt.hist(
    marks, 
    bins=bin_edges, 
    color='#2b7bba', 
    edgecolor='#599ed4', 
    rwidth=0.9
)

# 5. Add value labels on top of each bar for immediate data consumption
for count, bin_left in zip(counts, bins[:-1]):
    if count > 0:  # Only label bars that contain data
        plt.text(
            bin_left + 10, # Center of a 20-width bin
            count + 0.1,   # Offset slightly above the bar
            int(count), 
            ha='center', 
            va='bottom', 
            fontsize=11, 
            fontweight='bold', 
            color='#333333'
        )

# 6. Perfect the axis limits, ticks, and layout aesthetics
plt.xlim(0, 100)
plt.xticks(bin_edges)
plt.ylim(0, max(counts) + 0.8) # Padding for value labels

# 7. Use distinct font sizes and weights for titles and labels
plt.title('Distribution of Student Marks', fontsize=16, fontweight='bold', pad=15, color='#222222')
plt.xlabel('Marks (Score Range)', fontsize=12, labelpad=10, color='#444444')
plt.ylabel('Number of Students (Frequency)', fontsize=12, labelpad=10, color='#444444')

# 8. Mute background grid visibility so it stays secondary to the data bars
plt.grid(axis='x', linestyle='') # Remove vertical gridlines
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

#### **Scatter Plot**

```python
import matplotlib.pyplot as plt

## 1. Define the dataset
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 55, 60, 65, 70, 75, 80, 85]

plt.scatter(hours_studied, exam_scores, color='purple', edgecolor='black', s=100, alpha=0.7, label='Student Data Points', marker='o')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Hours Studied vs. Exam Scores')
plt.legend()
plt.show()
```

```python
## Including a Regression Line in the Scatter Plot
import matplotlib.pyplot as plt

hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 55, 60, 65, 70, 75, 80, 85]

plt.scatter(hours_studied, exam_scores, color='purple', edgecolor='black', s=100, alpha=0.7, label='Student Data Points', marker='o')
plt.plot(hours_studied, exam_scores, color='orange', linewidth=2, label='Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Hours Studied vs. Exam Scores with Regression Line')
plt.legend()
plt.show()
```

```python
## Comparing Two Datasets in a Scatter Plot
import matplotlib.pyplot as plt

dataset1_x = [1, 2, 3, 4, 5]
dataset1_y = [2, 3, 5, 7, 11]

dataset2_x = [1, 2, 3, 4, 5]
dataset2_y = [1, 4, 6, 8, 10]

plt.scatter(dataset1_x, dataset1_y, color='blue', label='Dataset 1', marker='o', s=100)
plt.scatter(dataset2_x, dataset2_y, color='red', label='Dataset 2', marker='^', s=100)
plt.plot(dataset1_x, dataset1_y, color='blue', linewidth=2, linestyle='--', label='Dataset 1 Trend')
plt.plot(dataset2_x, dataset2_y, color='red', linewidth=2, linestyle='--', label='Dataset 2 Trend')
plt.xlabel('X-axis Values')
plt.ylabel('Y-axis Values')
plt.title('Comparison of Two Datasets in a Scatter Plot')
plt.legend()
plt.show()
```
