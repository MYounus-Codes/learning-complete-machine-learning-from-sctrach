### ***Subplots and Layout Adjusment `(ADVANCE)`***

```python
## Making Charts Using Subplots

import matplotlib.pyplot as plt

# Create some sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a figure and a set of subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))  # 2 rows, 2 columns

# Plot data on each subplot
axs[0, 0].plot(x, y, 'r-', label='Line Plot')  # Red line
axs[0, 0].set_title('Red Line')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].grid(True)  # Add grid to the first subplot
axs[0, 0].legend()  # Add legend to the first subplot

axs[0, 1].scatter(x, y, color='b', label='Scatter Plot')  # Blue scatter
axs[0, 1].set_title('Blue Scatter')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].grid(True)  # Add grid to the first subplot
axs[0, 1].legend()  # Add legend to the first subplot


axs[1, 0].bar(x, y, color='g', edgecolor='black', linewidth=1.5, alpha=0.7, linestyle='--', hatch='/', width=0.5, label='Sales Data')  # Green bar chart
axs[1, 0].set_title('Green Bar Chart')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].grid(True)  # Add grid to the first subplot
axs[1, 0].legend()  # Add legend to the first subplot

axs[1, 1].hist(y, bins=5, color='y', label='Frequency Distribution')  # Yellow histogram
axs[1, 1].set_title('Yellow Histogram')
axs[1, 1].set_xlabel('Bins')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].grid(True)  # Add grid to the first subplot
axs[1, 1].legend()  # Add legend to the first subplot

# Adjust layout to prevent overlap
plt.tight_layout()

# Add a main title for the entire figure
fig.suptitle('Various Plots in Subplots', fontsize=16)  # Add a main title

# Show the plots
plt.show()
```

### ***Code Improved***

```python
import matplotlib.pyplot as plt

# -----------------------------
# Sample Data
# -----------------------------
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [20, 35, 50, 70, 110]

# Create Figure
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# =====================================================
# 1. Line Plot
# =====================================================

axs[0, 0].plot(
    months,
    sales,
    color="royalblue",
    linewidth=3,
    marker="o",
    markersize=8,
    markerfacecolor="white",
    markeredgewidth=2,
    linestyle="-",
    label="Monthly Sales"
)

axs[0, 0].set_title("Monthly Sales Trend", fontsize=14, fontweight="bold")
axs[0, 0].set_xlabel("Month", fontsize=11)
axs[0, 0].set_ylabel("Sales (Thousands)", fontsize=11)

axs[0, 0].grid(True, linestyle="--", alpha=0.4)
axs[0, 0].legend(loc="upper left")

# =====================================================
# 2. Scatter Plot
# =====================================================

axs[0, 1].scatter(
    months,
    sales,
    s=140,
    color="crimson",
    edgecolor="black",
    alpha=0.8,
    label="Sales Data"
)

axs[0, 1].set_title("Sales Distribution", fontsize=14, fontweight="bold")
axs[0, 1].set_xlabel("Month")
axs[0, 1].set_ylabel("Sales (Thousands)")

axs[0, 1].grid(True, linestyle=":", alpha=0.5)
axs[0, 1].legend()

# =====================================================
# 3. Bar Chart
# =====================================================

bars = axs[1, 0].bar(
    months,
    sales,
    color="mediumseagreen",
    edgecolor="black",
    linewidth=1.2,
    width=0.6,
    hatch="//",
    alpha=0.85,
    label="Sales"
)

# Add value labels on bars
axs[1, 0].bar_label(
    bars,
    padding=3,
    fontsize=10,
    color="black"
)

axs[1, 0].set_title("Monthly Sales Comparison", fontsize=14, fontweight="bold")
axs[1, 0].set_xlabel("Month")
axs[1, 0].set_ylabel("Sales (Thousands)")

axs[1, 0].grid(axis="y", linestyle="--", alpha=0.4)
axs[1, 0].legend()

# =====================================================
# 4. Histogram
# =====================================================

axs[1, 1].hist(
    sales,
    bins=6,
    color="gold",
    edgecolor="black",
    linewidth=1.2,
    alpha=0.8,
    label="Sales Frequency"
)

axs[1, 1].set_title("Distribution of Sales", fontsize=14, fontweight="bold")
axs[1, 1].set_xlabel("Sales Range")
axs[1, 1].set_ylabel("Frequency")

axs[1, 1].grid(True, linestyle="--", alpha=0.4)
axs[1, 1].legend()

# =====================================================
# Overall Figure
# =====================================================

fig.suptitle(
    "Monthly Sales Dashboard",
    fontsize=20,
    fontweight="bold"
)

plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()

## Saving the Figure

fig.savefig("monthly_sales_dashboard.png", dpi=300, bbox_inches="tight")
```
