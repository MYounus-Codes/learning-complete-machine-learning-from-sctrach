```python
## Making Charts Using Subplots

import matplotlib.pyplot as plt

# Create some sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a figure and a set of subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))  # 2 rows, 2 columns

# Plot data on each subplot
axs[0, 0].plot(x, y, 'r-', label='Line Plot')  # Red line
axs[0, 0].set_title('Red Line')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].grid(True)  # Add grid to the first subplot
axs[0, 0].legend()  # Add legend to the first subplot

axs[0, 1].scatter(x, y, color='b', label='Scatter Plot')  # Blue scatter
axs[0, 1].set_title('Blue Scatter')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].grid(True)  # Add grid to the first subplot
axs[0, 1].legend()  # Add legend to the first subplot


axs[1, 0].bar(x, y, color='g', edgecolor='black', linewidth=1.5, alpha=0.7, linestyle='--', hatch='/', width=0.5, label='Sales Data')  # Green bar chart
axs[1, 0].set_title('Green Bar Chart')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].grid(True)  # Add grid to the first subplot
axs[1, 0].legend()  # Add legend to the first subplot

axs[1, 1].hist(y, bins=5, color='y', label='Frequency Distribution')  # Yellow histogram
axs[1, 1].set_title('Yellow Histogram')
axs[1, 1].set_xlabel('Bins')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].grid(True)  # Add grid to the first subplot
axs[1, 1].legend()  # Add legend to the first subplot

# Adjust layout to prevent overlap
plt.tight_layout()

# Add a main title for the entire figure
fig.suptitle('Various Plots in Subplots', fontsize=16)  # Add a main title

# Show the plots
plt.show()
```
