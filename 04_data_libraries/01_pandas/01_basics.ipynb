```python
import pandas as pd
import numpy as np

# Create a simple sales dataset
data = {
    'Transaction_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110], 
    'Date': ['2026-07-01', '2026-07-01', '2026-07-02', '2026-07-02', '2026-07-03', 
             '2026-07-03', '2026-07-04', '2026-07-04', '2026-07-05', '2026-07-05'],
    'Product': ['Laptop', 'Mouse', 'Monitor', 'Laptop', 'Mouse', 
                'Keyboard', 'Monitor', 'Laptop', 'Mouse', np.nan], # np.nan introduces a missing value for practice
    'Category': ['Electronics', 'Accessories', 'Electronics', 'Electronics', 'Accessories', 
                 'Accessories', 'Electronics', 'Electronics', 'Accessories', 'Accessories'],
    'Units_Sold': [1, 4, 2, 1, 5, 2, np.nan, 3, 10, 1], # nan for data cleaning practice
    'Unit_Price': [1200, 25, 300, 1200, 25, 50, 300, 1200, 25, 45],
    'Region': ['North', 'East', 'West', 'South', 'East', 'North', 'West', 'North', 'South', 'East']
}

## Creating the DataFrame from the dictionary
df = pd.DataFrame(data)
print(df)
```

```python
df.head(5)  # Display the first 5 rows of the DataFrame
```

```python
df.info() # Display information about the DataFrame, including data types and non-null counts
```

```python
df.describe() # Display summary statistics for numerical columns in the DataFrame
```

```python
df["Category"].unique()  # Display unique values in the 'Category' column
```

```python
df.shape # Display the shape of the DataFrame (rows, columns)
```

```python
for col in df.columns:
    print(f"Column: {col}, Unique Values: {df[col].nunique()}")  # Display the number of unique values in each column
```

```python
df.isnull().sum()  # Count the number of missing values in each column
```

```python
# Printing the null value of Product colunm

df[df['Product'].isnull()]  # Display rows where 'Product' is null
```

```python
# Filling Null values

df["Product"] = df["Product"].fillna("Unknown", inplace=True)  # Fill null values in the 'Product' column with "Unknown"
print("Missing values in 'Product' column:", df["Product"].isnull().sum())  # Verify that there are no more null values in the 'Product' column
print("\n\n", df["Product"])
```

```python
df.to_csv("sales_data.csv", index=False)  # Save the DataFrame to a CSV file without the index
```
