```python
## Dropping the columns which are not required for analysis

df = df.drop(columns=["market_demand_score", "processor_score"]) 
df
```

```python
## Converting the all categorical columns to lowercase to maintain uniformity

for col in df.columns:
    if df[col].dtype == "str":  # Check if the column is of object type (categorical)
        df[col] = df[col].str.lower()  # Convert the values in the column to lowercase
```

```python
df.head()
```

```python
df["brand"].unique()
```

```python
# ['realme', 'samsung', 'apple', 'google', 'oneplus', 'vivo', 'xiaomi']
brand_to_nums = {
    "realme": 0,
    "samsung": 1,
    "apple": 2,
    "google": 3,
    "oneplus": 4,
    "vivo": 5,
    "xiaomi": 6
}

df["brand"] = df["brand"].map(brand_to_nums)  # Map the brand names to numerical values using the dictionary
```

```python
df["brand"].head(5)
```

```python
df["brand"].value_counts()  # Count the occurrences of each unique value in the "brand" column
```

```python
df["brand"].unique()  # Display the unique values in the "brand" column after mapping to numerical values
```

```python
df.head()
```

```python
# Dropping the model column as it is not required for analysis

df = df.drop(columns=["model"])
df.head(5)
```

```python
df["os_type"].unique()  # Display the unique values in the "os_type" column
```

```python
os_type_to_nums = {
    "android": 0,
    "ios": 1
}

df["os_type"] = df["os_type"].map(os_type_to_nums)  # Map the OS types to numerical values using the dictionary
```

```python
df.head(5)
```

```python
df["city_tier"].unique()  # Display the unique values in the "city_tier" column
```

```python
city_tier_to_nums = {
    "tier1": 0,
    "tier2": 1,
    "tier3": 2
}

df["city_tier"] = df["city_tier"].map(city_tier_to_nums)  # Map the city tiers to numerical values using the dictionary
```

```python
df.head(5)  # Display the first 5 rows of the DataFrame
```

```python
df["seller_type"].unique()  # Display the unique values in the "seller_type" column
```

```python
df["seller_type"].isna().sum()  # Count the number of missing values in the "seller_type" column
```

```python
seller_type_to_nums = {
    "individual": 0,
    "store": 1
}

df["seller_type"] = df["seller_type"].map(seller_type_to_nums)  # Map the seller types to numerical values using the dictionary
```

```python
df.head(5)  # Display the first 5 rows of the DataFrame
```

```python
df.info()
```

```python
df["condition"].unique()  # Display the unique values in the "condition" column
```

```python
# ['excellent', 'fair', 'good', 'poor']
condition_to_nums = {
    "excellent": 0,
    "fair": 1,
    "good": 2,
    "poor": 3
}

df["condition"] = df["condition"].map(condition_to_nums)  # Map the condition values to numerical values using the dictionary
```
